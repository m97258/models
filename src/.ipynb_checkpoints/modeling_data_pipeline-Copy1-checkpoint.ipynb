# 3.- Data Transformation / Processing

### 3.1 Data Transformation / Processing: Splitting the dataset in Trainvalidation and Test datasets

```python
# A split of the data is made in a "trainvalidation" dataset and a "test" dataset.
percentage = 0.20
Xtrainval, Xtest, Ytrainval, Ytest = modeling_library.trainvalidation_testing_split(df_data, target_column, percentage)
```

### 3.2 Data Transformation/Processing: Data Standardization

```python
# I identify categorical and non-categorical variables, for giving them an ad-hoc standardization
cols_non_categorical = [age_column, number_of_children_colum, income_column]
cols_categorical     = [marital_status_column, education_level_column, smoking_status_column, physical_activity_level_column, \
                        employment_status_column, alcohol_consumption_column, dietary_habits_column, sleep_patterns_column, \
                        history_of_substance_abuse_column, family_history_of_depression_column, chronic_medical_conditions_column, \
                        predicted_gender_column, last_name_origin_column]
```

```python
# I standardize the trainvalidation dataset, getting dataframes of means and stdev for standardizing test anb holdout data.
Xtrainval_norm, cols_dummy, cols_all_variables, dfmean, dfstd = modeling_library.standardizing_dataset_modeling(Xtrainval, cols_categorical, \
                                                                                                                cols_non_categorical)
Ytrainval_norm = Ytrainval.replace(['No', 'Yes'],[0, 1])

# I standardize the Test dataset with the mean and stdev found standardizing the trainvalidation dataset
Xtest_norm     = data_pipelines_library.standardizing_dataset_production(Xtest, cols_categorical, cols_non_categorical, cols_dummy, \
                                                                        cols_all_variables, dfmean, dfstd)
Ytest_norm     = Ytest.replace(['No', 'Yes'],[0, 1])
```

### 3.3 Data Transformation / Processing: Removing features

#### Removing blatantly non predictive variables

```python
# Removing categorical variables whose values showed frequencies of illness that differed no more than 0.1% in 
# regard to the average frequency of illness for the whole population
columns_to_be_removed_non_predictive = ['History of Substance Abuse_No', 'History of Substance Abuse_Yes', \
                                        'Predicted Gender_female', 'Predicted Gender_male', 'Predicted Gender_unknown', \
                                        'Last Name Origin_Asian', 'Last Name Origin_Western European', 'Last Name Origin_Latin European']
cols_categorical.remove(history_of_substance_abuse_column)
cols_categorical.remove(predicted_gender_column)
cols_categorical.remove(last_name_origin_column)
for col in columns_to_be_removed_non_predictive:
    if (col in cols_dummy):
        cols_dummy.remove(col)
Xtrainval_norm = data_pipelines_library.remove_discarded_columns(Xtrainval_norm, columns_to_be_removed_non_predictive)
Xtest_norm     = data_pipelines_library.remove_discarded_columns(Xtest_norm,     columns_to_be_removed_non_predictive)
```

#### Removing linear colinearities (variables that are a linear combination of other variables)

```python
# In every categorical variable, the dummy variables creation procedure has created as many dummy variables as the number
# of vuales a categorical variable had. Sstematically on of those variables was determined by the rest. For instance:
# 'Marital Status_Married' = NOT('Marital Status_Divorced' OR 'Marital Status_Single' OR 'Marital Status_Widowed')
# between every set of dummy variables that are entangled, I discarded the one that had an observed frequency of mental
# illness the closest possible to the observed average for the population.
columns_to_be_removed_multicolinearities = ['Marital Status_Divorced', 'Education Level_Master\'s Degree', 'Smoking Status_Former', \
                                            'Physical Activity Level_Moderate', 'Employment Status_Employed', 'Alcohol Consumption_Low', \
                                            'Dietary Habits_Moderate', 'Sleep Patterns_Fair', 'Family History of Depression_No', \
                                            'Chronic Medical Conditions_No']
for col in columns_to_be_removed_multicolinearities:
    if (col in cols_dummy):
        cols_dummy.remove(col)
Xtrainval_norm = data_pipelines_library.remove_discarded_columns(Xtrainval_norm, columns_to_be_removed_multicolinearities)
Xtest_norm     = data_pipelines_library.remove_discarded_columns(Xtest_norm,     columns_to_be_removed_multicolinearities)
```

#### Removing strongly correlated variables

```python
# Correlation matrix of all the eventual explanatory features
visualization_library.plot_correlations(Xtrainval_norm)
```

```python
# Pairs of variables with the highest correlation in absolute value
visualization_library.highlight_highest_abs_correlations(Xtrainval_norm , threshold_percentage=0.6)
```

```python
# Removing variables highly correlated (in absolute terms)
columns_to_be_removed_highly_correlated = ['Marital Status_Single', 'Employment Status_Unemployed']
for col in columns_to_be_removed_highly_correlated:
    if (col in cols_dummy):
        cols_dummy.remove(col)
Xtrainval_norm = data_pipelines_library.remove_discarded_columns(Xtrainval_norm, columns_to_be_removed_highly_correlated)
Xtest_norm     = data_pipelines_library.remove_discarded_columns(Xtest_norm,     columns_to_be_removed_highly_correlated)
```

#### Removing variables with a low feature importance in preliminary model

```python
# A preliminary model is developed using standard parameters
modelXGBoost_preliminary = XGBClassifier(colsample_bytree = 0.6, learning_rate = 0.1, max_depth = 6, n_estimators = 1000, verbosity = 0)
modelXGBoost_preliminary.fit(Xtrainval_norm, Ytrainval_norm)
```

```python
# The feature importances of that model are calculated
feature_importances = pd.DataFrame(modelXGBoost_preliminary.feature_importances_, Xtest_norm.columns, columns = ['importance'])
feature_importances = feature_importances.sort_values(by='importance', ascending=False)
ranked_features =  list(feature_importances.index)
feature_importances.head(10)
```

```python
# Removing features with low feature importance
columns_to_be_removed_low_feature_importance = ranked_features[10:]
for col in columns_to_be_removed_low_feature_importance:
    if (col in cols_non_categorical):
        cols_non_categorical.remove(col)
        dfstd  = dfstd.drop(dfstd[dfstd[variables_column]==col].index)
        dfmean = dfmean.drop(dfmean[dfmean[variables_column]==col].index)
for col in columns_to_be_removed_low_feature_importance:
    if (col in cols_dummy):
        cols_dummy.remove(col)
Xtrainval_norm = data_pipelines_library.remove_discarded_columns(Xtrainval_norm, columns_to_be_removed_low_feature_importance)
Xtest_norm     = data_pipelines_library.remove_discarded_columns(Xtest_norm,     columns_to_be_removed_low_feature_importance)
```

# 4.- Data Ingestion

```python
# The data on which the model would be used is properly stored
data_pipelines_library.storing_dataset(Xtrainval_norm, name_path_tests, default_name_file_Xtrain_modeling)
data_pipelines_library.storing_dataset(pd.DataFrame(Ytrainval_norm), name_path_tests, default_name_file_Ytrain_modeling)
```

# 5.- Modeling

### 5.1 Modeling: building a series of Gradient Boosting Machine models on a cross-validation framework

```python
# Getting the best parameters for the XGBOOST model, looking at the maximum average gini coefficient on the validation dataset
results_grid_search   = modeling_library.gridsearch_XGBoost(Xtrainval_norm, Ytrainval_norm)
Best_colsample_bytree = results_grid_search['col_sample_bytree']
Best_learning_rate    = results_grid_search['learning_rate']
Best_max_depth        = results_grid_search['max_depth']
Best_n_estimators     = results_grid_search['n_estimators']
```

```python
print(" Model parameters - XGBoost model:")
print(" - col sample by tree = ", Best_colsample_bytree)
print(" - learning rate      = ", Best_learning_rate)
print(" - max depth          = ", Best_max_depth)
print(" - n estimators       = ", Best_n_estimators)
```

### 5.2 Modeling: a model is retrained on the whole trainvalidation data, using parameters found at crossvalidation.

```python
modelXGBoost = XGBClassifier(colsample_bytree = Best_colsample_bytree, learning_rate = Best_learning_rate, \
                             max_depth = Best_max_depth, n_estimators = Best_n_estimators, verbosity = 0)
modelXGBoost.fit(Xtrainval_norm, Ytrainval_norm)
```

# 6.- Model Evaluation

### 6.1 Model Evaluation: Getting results of Gradient Boosting Machine model on the Test dataset

```python
y_predtest_second_model      = modelXGBoost.predict(Xtest_norm)
Y_probabilities_second_model = modelXGBoost.predict_proba(Xtest_norm)[:,1]
```

### 6.2 Model Evaluation: Getting Gradient Boosting Machine performance

```python
# Plotting model performance across training, validation and test datasets
title          = ' Performance of GBM model on training, validation and testing datasets \n'
labels_titles  = ['Performance metrics', 'Accuracy Score']
avg_accuracy_train = results_grid_search['avg_accuracy_train']
avg_accuracy_val   = results_grid_search['avg_accuracy_val']
accuracy_test  = performance_library.accuracy_score(Ytest_norm, y_predtest_second_model) 
visualization_library.plot_performance_train_val_test(avg_accuracy_train, avg_accuracy_val, accuracy_test, title, labels_titles)
```

```python
# Getting predictions of the model with standard threshold of 50%
title = 'Performance metrics of GBM model for depression with threshold of 50% \n'
labels_titles = ['Performance metrics', 'Score']
visualization_library.plot_model_performance(Ytest_norm, y_predtest_second_model, title, labels_titles)
```
